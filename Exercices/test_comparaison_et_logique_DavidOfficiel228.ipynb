```python
# ğŸ“ ComplÃ©tez les fonctions suivantes en utilisant des opÃ©rateurs de comparaison et logiques

def est_entre(val: int, borne_min: int, borne_max: int) -> bool:
    """
    Retourne True si val est strictement entre borne_min et borne_max.
    """
    if borne_min < val < borne_max:
        return True
    else:
        return False  

def sont_egaux_ou_negatifs(a: int, b: int) -> bool:
    """
    Retourne True si a et b sont Ã©gaux OU si les deux sont nÃ©gatifs.
    """
    if a == b or (a < 0 and b < 0):
        return True 
    else:
        return False

def a_un_age_valide(age: int) -> bool:
    """
    Retourne True si lâ€™Ã¢ge est entre 18 et 120 inclus.
    """
    if 18 <= age <= 120:
        return True
    else:
        return False

def mot_de_passe_valide(mot_de_passe: str) -> bool:
    """
    Retourne True si le mot de passe fait au moins 8 caractÃ¨res et ne contient pas dâ€™espace.
    """
    if len(mot_de_passe) >= 8 and ' ' not in mot_de_passe:
        return True
    else:
        return False

def est_non_vide_et_non_null(texte: str) -> bool:
    """
    Retourne True si le texte nâ€™est ni vide ni Ã©gal Ã  'null' ou 'None' (en tant que chaÃ®nes).
    """
    if texte and texte.lower() not in ['null', 'none']:
        return True
    else:
        return False     
```
