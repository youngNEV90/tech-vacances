```python
# ğŸ“ ComplÃ©tez les fonctions suivantes pour travailler avec des listes

def ajouter_element(liste: list[int], element: int) -> list[int]:
    """
    Ajoute un Ã©lÃ©ment Ã  la fin de la liste et retourne la nouvelle liste.
    """
    liste.append(element)
    return liste

def supprimer_element(liste: list[int], element: int) -> list[int]:
    """
    Supprime la premiÃ¨re occurrence de l'Ã©lÃ©ment dans la liste, si elle existe.
    Retourne la liste modifiÃ©e.
    """
    if element in liste:
        liste.remove(element)
    return liste
        

def trier_liste(liste: list[int]) -> list[int]:
    """
    Retourne une nouvelle liste triÃ©e sans modifier lâ€™originale.
    """
    return sorted(liste)

def contient_element(liste: list[int], element: int) -> bool:
    """
    Retourne True si l'Ã©lÃ©ment est prÃ©sent dans la liste, False sinon.
    """
    if element in liste:
        return True
    else:
        return False

def moyenne_liste(liste: list[int]) -> float:
    """
    Retourne la moyenne des Ã©lÃ©ments de la liste.
    Si la liste est vide, retourne 0.0.
    """
    if not liste:
        return 0.0
    return sum(liste) / len(liste)
```

```python
# âœ… Cette cellule vÃ©rifie la validitÃ© des fonctions dÃ©finies ci-dessus

def run_list_tests():
    tests = [
        ("ajouter_element", ajouter_element, ([1, 2], 3), [1, 2, 3]),
        ("ajouter_element", ajouter_element, ([], 5), [5]),

        ("supprimer_element", supprimer_element, ([1, 2, 3], 2), [1, 3]),
        ("supprimer_element", supprimer_element, ([1, 2, 2], 2), [1, 2]),
        ("supprimer_element", supprimer_element, ([1, 2, 3], 4), [1, 2, 3]),  # Ã‰lÃ©ment non prÃ©sent

        ("trier_liste", trier_liste, ([3, 1, 2],), [1, 2, 3]),
        ("trier_liste", trier_liste, ([5, 5, 2],), [2, 5, 5]),

        ("contient_element", contient_element, ([1, 2, 3], 2), True),
        ("contient_element", contient_element, ([1, 2, 3], 5), False),

        ("moyenne_liste", moyenne_liste, ([1, 2, 3],), 2.0),
        ("moyenne_liste", moyenne_liste, ([10],), 10.0),
        ("moyenne_liste", moyenne_liste, ([],), 0.0),
    ]

    passed = 0
    failed = 0

    for name, func, inputs, expected in tests:
        try:
            result = func(*inputs)
            if result == expected:
                print(f"âœ… {name}{inputs} == {expected} â†’ OK")
                passed += 1
            else:
                print(f"âŒ {name}{inputs} â†’ {result}, attendu : {expected}")
                failed += 1
        except Exception as e:
            print(f"âŒ {name}{inputs} â†’ Erreur inattendue : {e}")
            failed += 1

    print("\nğŸ“Š RÃ©sumÃ© des tests :")
    print(f"âœ… RÃ©ussis : {passed}")
    print(f"âŒ Ã‰chouÃ©s : {failed}")

run_list_tests()
```
